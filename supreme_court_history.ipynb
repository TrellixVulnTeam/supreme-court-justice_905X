```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
os.listdir()
r='#FF9999'
b= '#759EF0'
red = 'background-color: '+r
blue = 'background-color: '+b


def change_color_group(_):
    df = _.copy()
    df.loc[df['Nomination By Party'] == 'R', :] = red
    df.loc[df['Nomination By Party'] == 'D', :] = blue
    return df  
```

```python
df = pd.read_csv(os.listdir()[1], header=0)
#df = df.drop('0', axis=1)
df['Total Appointments']=1
df['Result & Date***'] = pd.to_datetime(df['Result & Date***'], format='%m/%d/%Y')
df['Nominated*'] = pd.to_datetime(df['Nominated*'], format='%m/%d/%Y')

df['Senate-President'] = ''
df.loc[(df['Senate Majority']=='R') & (df['Nomination By Party']=='R'), 'Senate-President'] = 'R-R'
df.loc[(df['Senate Majority']=='D') & (df['Nomination By Party']=='D'), 'Senate-President'] = 'D-D'
df.loc[(df['Senate Majority']=='R') & (df['Nomination By Party']=='D'), 'Senate-President'] = 'R-D'
df.loc[(df['Senate Majority']=='D') & (df['Nomination By Party']=='R'), 'Senate-President'] = 'D-R'

df['Served'] = np.where(df['Result']=='C', "Yes", "No")

election_years = [i for i in range(2020, 1890, -4)]
df['Election Year'] = df['Nominated*'].apply(lambda d: d.year in election_years)
```

### Presidents with most successful nominations since 1900

```python
keep_cols = ['Nominated By', 'Total Appointments']
_ = df[df['Result']=='C'][keep_cols]
p = _.groupby(['Nominated By']).count().sort_values(by='Total Appointments', ascending=False)
p = pd.merge(p, df[['Nominated By', 'Nomination By Party']].drop_duplicates(), how='left', left_index=True, right_on='Nominated By')
p.style.apply(change_color_group, axis=None)

```

### Confirmations by party successful nomination

```python
keep_cols = ['Nomination By Party', 'Total Appointments']
_ = df[df['Result']=='C'][keep_cols]
p = _.groupby(['Nomination By Party']).count().sort_values(by='Total Appointments', ascending=False)
plt.barh(p.index, p['Total Appointments'], color=[r, b])
plt.title('Confirmations by party since 1900')
plt.xlabel('Number of Supreme Court Confirmations')
plt.ylabel('President Party Affiliation')
plt.show()
```

```python
keep_cols = ['Nomination By Party', 'Total Appointments']
_ = df[df['Result']=='C']
_ = _[_['Nominated*']>'1970-01-01'][keep_cols]
p = _.groupby(['Nomination By Party']).count().sort_values(by='Total Appointments', ascending=False)
plt.barh(p.index, p['Total Appointments'], color=[r, b])
plt.title('Confirmations by party since 1970')
plt.xlabel('Number of Supreme Court Confirmations')
plt.ylabel('President Party Affiliation')
plt.show()
```
